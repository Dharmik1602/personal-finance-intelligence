```python
import pandas as pd
import plotly.express as px
```

```python
# Load the cleaned data from Module 4
df = pd.read_csv('cleaned_expenses.csv')
df['date'] = pd.to_datetime(df['date'])
```

```python
# --- 1. Category-wise Spending (Pie Chart) ---
category_totals = df.groupby('category')['amount'].sum().reset_index()
fig_pie = px.pie(category_totals, values='amount', names='category', 
                 title='Where is the Money Going?', hole=0.4)
fig_pie.show()
```

```python
# --- 2. Monthly Spending Trend ---
monthly_trend = df.groupby('month')['amount'].sum().reset_index()
fig_line = px.line(monthly_trend, x='month', y='amount', 
                  title='Total Monthly Spending Trend', markers=True)
fig_line.show()
```

```python
# --- 3. Weekend vs Weekday Behavior ---
weekend_analysis = df.groupby('is_weekend')['amount'].mean().reset_index()
weekend_analysis['day_type'] = weekend_analysis['is_weekend'].map({True: 'Weekend', False: 'Weekday'})
fig_bar = px.bar(weekend_analysis, x='day_type', y='amount', 
                 title='Average Spending: Weekdays vs Weekends', color='day_type')
fig_bar.show()
```
